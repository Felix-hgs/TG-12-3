```python
import math
import numpy as np
class Erdbeschleunigung:
    def __init__(self, g=9.81, h=0 ,M=10):
        self.g = g
        self.h = h
        self.M = M

    def beschleunigung(self):
        return self.g
    
    def hoehe(self):
        return self.h
    
    def Erggeschwindigkeit(self):
        return (2 * self.g * self.h) ** 0.5
    
    def Meter(self):
        return self.M
    
    def __str__(self):

        for h in range(self.M):
            v = math.sqrt(2 * self.g * h)
            print(f"Bei {h} Meter Hoehe ist die Geschwindigkeit {v:.5f} m/s")
    
g = Erdbeschleunigung(h=10, M=11)
g.__str__()

x=np.arange(-10,11)
print(x)
y=x**2
print(y)

from matplotlib import pyplot as plt
plt.plot(x,y)
plt.xlabel("x-Achse")
plt.ylabel("y-Achse") 
plt.grid()
plt.xticks(np.arange(-10,11,1)) # Ticks alle 1
plt.show()

```
