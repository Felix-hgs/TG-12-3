```python
import numpy as np
from matplotlib import pyplot as plt
class Ball_Fall:
    def __init__(self, height, g=9.81, t_step=1):
        self.height = height
        self.g = g
        self.t_step = t_step

    def hoehe_und_zeit(self):
        t_total = (2 * self.height / self.g) ** 0.5
        t = np.arange(0, t_total + self.t_step, self.t_step)
        h = self.height - 0.5 * self.g * t**2
        h = np.maximum(h, 0)
        # Geschwindigkeit nur solange h > 0, danach 0
        v = self.g * t
        v = np.where(h > 0, v, 0)
        return t, h, v

    def fall_time(self):
        t = 0
        h = self.height
        print(f"Start: Höhe = {h:.2f} m, Geschwindigkeit = 0.00 m/s")
        while h > 0:
            v = self.g * t
            h = self.height - 0.5 * self.g * t**2
            h = max(h, 0)
            v = v if h > 0 else 0
            print(f"Nach {t:.2f} s: Höhe = {h:.2f} m, Geschwindigkeit = {v:.2f} m/s")
            t += self.t_step
        time_total = (2 * self.height / self.g) ** 0.5
        print(f"Gesamte Fallzeit: {time_total:.2f} s")
        return time_total

a = Ball_Fall(100, t_step=0.5)  # Beispiel: t_step=0.5 für halbe Sekunden

a.fall_time()

t, h, v = a.hoehe_und_zeit()

plt.plot(t, h, label="Höhe (m)")
plt.plot(t, v, label="Geschwindigkeit (m/s)")
plt.xlabel("Zeit (s)")
plt.ylabel("Höhe (m) / Geschwindigkeit (m/s)")
plt.xticks(np.arange(0,6,0.5)) # Ticks alle 1
plt.legend()
plt.grid()
plt.show()
plt.pie([1, 2, 3], labels=["A", "B", "C"])


```
