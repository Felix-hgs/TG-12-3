```python
import numpy as np
import matplotlib.pyplot as plt

tend = 10
dt = 0.1
t = 0
v = 0
s = 0
g = 9.81
h0 = 0
p = int(tend//dt)

strecke=[]
zeit=[]
geschwindigkeit=[]
streckea=[]

for i in range (1,p):
    t += dt
    a = g
    v = v + a * dt
    s = s + v * dt
    ht = h0 + 0.5 * a * t**2
    vt = g * t
    print (f"t={t:5.2f} s, v={v:8.3f} m/s, s={s:8.3f} m",f" ht={ht:8.3f} m",f" vt={vt:8.3f} m/s")
    streckea.append(ht)
    strecke.append(s)
    zeit.append(t)
    geschwindigkeit.append(v)
    





fig=plt.figure()
ax1=fig.add_subplot(2,1,1)
ax2=fig.add_subplot(2,1,2)
ax1.set_ylabel('geschwindigkeit in m/s')
ax1.set_xlabel('zeit in s')
ax1.grid(True)
ax2.set_xlabel('zeit in s')
ax2.set_ylabel('strecke in m')
ax2.grid(True)
ax1.plot(zeit,geschwindigkeit,label='f(x)',color='red')
ax2.plot(zeit,strecke,label='g(x)',color='blue')
ax2.plot(zeit,streckea,label='g(x)',color='green')
ax1.legend()
ax2.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

tend = 10
dt = 0.1
t = 0
v = 0
s = 0
g = 9.81
h0 = 0
cw = 1.11
A = 40
pl = 1.23
p = int(tend//dt)

strecke=[]
zeit=[]
geschwindigkeit=[]
streckea=[]

for i in range (1,p):
    a = g - (cw * A * pl * v**2) / (2 * 80)
    t += dt
    v = v + a * dt
    s = s + v * dt
    ht = h0 + 0.5 * a * t**2
    vt = g * t
    print (f"t={t:5.2f} s, v={v:8.3f} m/s, s={s:8.3f} m",f" ht={ht:8.3f} m",f" vt={vt:8.3f} m/s")
    streckea.append(ht)
    strecke.append(s)
    zeit.append(t)
    geschwindigkeit.append(v)
    





fig=plt.figure()
ax1=fig.add_subplot(2,1,1)
ax2=fig.add_subplot(2,1,2)
ax1.set_ylabel('geschwindigkeit in m/s')
ax1.set_xlabel('zeit in s')
ax1.grid(True)
ax2.set_xlabel('zeit in s')
ax2.set_ylabel('strecke in m')
ax2.grid(True)
ax1.plot(zeit,geschwindigkeit,label='f(x)',color='red')
ax2.plot(zeit,strecke,label='g(x)',color='blue')
ax2.plot(zeit,streckea,label='g(x)',color='green')
ax1.legend()
ax2.legend()
plt.show()
```
